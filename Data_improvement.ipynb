```python
# SQL query to get 10 rows with NULL values in specified columns
sql_query = """
SELECT *
FROM movies_data
WHERE director IS NULL 
   OR cast IS NULL 
   OR country IS NULL 
   OR date_added IS NULL 
   OR rating IS NULL 
   OR duration IS NULL
LIMIT 10;
"""

# You can execute this query using your existing MySQL connection:
cursor.execute(sql_query)
results = cursor.fetchall()

# Convert results to a pandas DataFrame for better visualization
columns = [desc[0] for desc in cursor.description]
df_null_analysis = pd.DataFrame(results, columns=columns)

# Display the results
display(df_null_analysis)
```

```python
# Execute SQL command to get distinct entries in the 'type' column
cursor.execute("SELECT DISTINCT type FROM movies_data")

# Fetch all results
distinct_types = cursor.fetchall()

# Display the distinct values
print("Distinct entries in 'type' column:")
for entry in distinct_types:
    print(entry[0])  # Print each distinct type
```
